**Question 7**

Statistically test any variable pairs with a correlation coefficient > |0.1| for its difference from 0 and report your result. Is the direction and significance of your finding as expected?



**Answer:**

Yes, as expected and as it is shown from the table that the two variables education-num	, hours-per-week has a very weak positive corrolation.

```python
from scipy.stats import ttest_ind


# Identify variable pairs with correlation coefficient > |0.1|
threshold = 0.1
pairs_to_test = []
for i in range(len(cols_for_corr)):
    for j in range(i + 1, len(cols_for_corr)):
        col1 = cols_for_corr[i]
        col2 = cols_for_corr[j]
        correlation = correlation_matrix.loc[col1, col2]
        if abs(correlation) > threshold:
            pairs_to_test.append((col1, col2))

# Perform t-tests for significant differences from 0
results = []
for col1, col2 in pairs_to_test:
    # Handle potential missing values
    valid_data = adult_data[[col1, col2]].dropna()

    if len(valid_data) > 0:
      t_statistic, p_value = ttest_ind(valid_data[col1], valid_data[col2])
      results.append({
          'Variable 1': col1,
          'Variable 2': col2,
          'Correlation': correlation_matrix.loc[col1, col2],
          't-statistic': t_statistic,
          'p-value': p_value
      })
    else:
      results.append({
          'Variable 1': col1,
          'Variable 2': col2,
          'Correlation': correlation_matrix.loc[col1, col2],
          't-statistic': 'N/A (No valid data after dropping NaN)',
          'p-value': 'N/A (No valid data after dropping NaN)'
      })

# Report the results
results_df = pd.DataFrame(results)
results_df
```

**Question 8**

How does the correlation (and its significance) between education_num and age compare between male and female participants? Is this expected?


```python

# Strip leading/trailing whitespaces from the 'sex' column
adult_data['sex'] = adult_data['sex'].str.strip()
grouped_corr = adult_data.groupby('sex')[['education-num', 'age']].corr()

# Extract correlations for males and females
male_corr = grouped_corr.loc['Male', 'education-num']['age']
female_corr = grouped_corr.loc['Female', 'education-num']['age']

print(f"Correlation between education-num and age for males: {male_corr}")
print(f"Correlation between education-num and age for females: {female_corr}")

# Perform t-tests to compare the correlations
from scipy.stats import ttest_ind

# Separate data for males and females
male_data = adult_data[adult_data['sex'] == 'Male']
female_data = adult_data[adult_data['sex'] == 'Female']


# Perform t-test
t_statistic, p_value = ttest_ind(male_data['education-num'], female_data['education-num'])

print(f"\nT-test results comparing education-num between males and females:")
print(f"t-statistic: {t_statistic}")
print(f"p-value: {p_value}")


# Significance of the difference
alpha = 0.05  # Significance level
if p_value < alpha:
    print("The difference in correlations is statistically significant.")
else:
    print("The difference in correlations is not statistically significant.")
```

```python
import pandas as pd
import statsmodels.api as sm

# Load your dataset
adult_data = pd.read_csv('/content/adult.data', header=None,
                         names=['age', 'workclass', 'fnlwgt', 'education',
                                'education-num', 'marital-status', 'occupation',
                                'relationship', 'race', 'sex', 'capital-gain',
                                'capital-loss', 'hours-per-week', 'native-country',
                                'income'])

# Data preprocessing
adult_data['age'] = pd.to_numeric(adult_data['age'], errors='coerce')
adult_data['education-num'] = pd.to_numeric(adult_data['education-num'], errors='coerce')
adult_data['hours-per-week'] = pd.to_numeric(adult_data['hours-per-week'], errors='coerce')

# Convert 'income' to numerical (0 and 1) and handle NaN values
# Replace '?' with NaN for consistent handling of missing values
adult_data['income'] = adult_data['income'].replace(' ?', pd.NA).str.strip() # Replace '?' with NaN and remove leading/trailing spaces
adult_data['income'] = adult_data['income'].map({'<=50K': 0, '>50K': 1}).astype(float) # convert income to float

adult_data.dropna(subset=['age', 'education-num', 'hours-per-week', 'income'], inplace=True)

# Define independent (X) and dependent (y) variables
X = adult_data[['age', 'education-num', 'hours-per-week']]
y = adult_data['income']

# Add a constant to the model (intercept)
X = sm.add_constant(X)

# Fit the model
model = sm.Logit(y, X)
result = model.fit()

# Summary of the model
print(result.summary())

# Make predictions
predictions = result.predict(X)
print(predictions)
```

```python
# prompt: plot linear regression

import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'result' from your previous code contains the fitted model
# and 'X' contains the independent variables and 'y' the dependent variable.

# Create a scatter plot with the regression line
plt.figure(figsize=(10, 6))
sns.regplot(x=X['age'], y=y, data=adult_data, scatter_kws={'alpha': 0.5})
plt.xlabel('Age')
plt.ylabel('Income')  # Assuming your dependent variable is 'income'
plt.title('Linear Regression: Age vs. Income')
plt.show()

# You can create similar plots for 'education-num' and 'hours-per-week':
plt.figure(figsize=(10, 6))
sns.regplot(x=X['education-num'], y=y, data=adult_data, scatter_kws={'alpha': 0.5})
plt.xlabel('Education Number')
plt.ylabel('Income')
plt.title('Linear Regression: Education Number vs. Income')
plt.show()


plt.figure(figsize=(10, 6))
sns.regplot(x=X['hours-per-week'], y=y, data=adult_data, scatter_kws={'alpha': 0.5})
plt.xlabel('Hours per Week')
plt.ylabel('Income')
plt.title('Linear Regression: Hours per Week vs. Income')
plt.show()
```

```python
# prompt: linear regression hours_per_week as the dependent variable and sex as the independent variable

# Convert 'sex' to numerical (0 and 1)
adult_data['sex'] = adult_data['sex'].map({' Male': 0, ' Female': 1}).astype(int)

# Define independent (X) and dependent (y) variables
X = adult_data['sex']
y = adult_data['hours-per-week']

# Add a constant to the model (intercept)
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X)
results = model.fit()

# Print the model summary
print(results.summary())

# Create a scatter plot with the regression line
plt.figure(figsize=(8, 6))
sns.regplot(x='sex', y='hours-per-week', data=adult_data)
plt.xlabel('Sex (0: Male, 1: Female)')
plt.ylabel('Hours per Week')
plt.title('Linear Regression: Hours per Week vs. Sex')
plt.show()
```

The linear regression model examining the relationship between `hours-per-week` (dependent variable) and `sex` (independent variable) reveals a statistically significant relationship.  

The p-value associated with the sex coefficient is less than 0.05, indicating that the difference in average hours worked per week between males and females is unlikely due to random chance.  

The coefficient for sex suggests the difference in mean hours worked between the two groups.  


```python

# Group data by sex and calculate the mean hours per week
hours_by_sex = adult_data.groupby('sex')['hours-per-week'].mean()

# Print the results
print(hours_by_sex)

# Determine who works more hours on average
# Use numerical keys (0 and 1) instead of string keys (' Male' and ' Female')
if hours_by_sex[0] > hours_by_sex[1]:  # Changed from ' Male' to 0 and ' Female' to 1
    print("\nMen work more hours on average.")
elif hours_by_sex[0] < hours_by_sex[1]:  # Changed from ' Male' to 0 and ' Female' to 1
    print("\nWomen work more hours on average.")
else:
    print("\nMen and women work the same number of hours on average.")
```

```python

# Convert 'sex' to numerical (0 and 1) if not already done, handling NaNs
adult_data['sex'] = adult_data['sex'].map({' Male': 0, ' Female': 1}).fillna(-1).astype(int)
#fillna handles missing data by replacing it with -1

# Add 'education-num' as a control variable in the regression
import statsmodels.formula.api as smf

# Define the formula for the regression model including 'education-num' as a control variable
formula = 'Q("hours-per-week") ~ sex + Q("education-num")'

# Fit the model
model = smf.ols(formula, data=adult_data).fit()

# Print the model summary
print(model.summary())

# Check the p-value for education-num
print("\np-value for education_num:", model.pvalues['Q("education-num")'])

```
